```python
def find_circular_routes(df, departure):
  """
  This function finds all routes in the 'df' dataframe that 
  eventually lead back to the initial 'departure' point using iteration.

  Args:
      df: A pandas dataframe with 'account_id' and 'counterparty_id' columns (pd.DataFrame).
      departure: The initial departure point (string).

  Returns:
      A list of lists, where each inner list represents a circular route.
  """
  circular_routes = []
  visited = set()
  stack = [(departure, [])]  # Stack for iterative exploration

  while stack:
    current_point, route = stack.pop()

    if current_point == departure and len(route) > 1:  # Found a circular route
      circular_routes.append(route[:])  # Append a copy to avoid modifying original
      continue

    if current_point not in visited:
      visited.add(current_point)
      for index, row in df[df['account_id'] == current_point].iterrows():
        next_stop = row['counterparty_id']
        stack.append((next_stop, route + [next_stop]))

  return circular_routes

# Example usage
# Assuming 'df' is your pandas dataframe containing flight routes
departure_point = "U0IZMAXA2PYLDJNE"
circular_routes = find_circular_routes(df_final.copy(), departure_point)  # Avoid modifying original df

print(f"Circular routes starting from {departure_point}:")
for route in circular_routes:
  print(" -> ".join(route))
```
